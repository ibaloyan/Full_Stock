```python
year = 1980

x = 0
changes = []
while x < (len(Micodata)-1):
    openV = Micodata["Open"][x]

    closeV = Micodata["Close"][x+1]
    changes.append(round((closeV-openV)/openV*100,2))

    x += 1
print(len(changes))  
```
