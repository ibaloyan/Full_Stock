```python

year = 1980

x = 0
changes = []
while x < (len(Appledata)-1):
    openV = Appledata["Open"][x]

    closeV = Appledata["Close"][x+1]
    changes.append(round((closeV-openV)/openV*100))

    x += 1
print(len(changes))    

```

```python
months = ["Jan-Feb","Feb-Mar","Mar-Apr","Apr-May","May-June","June-July","July-Aug","Aug-Sept","Sept-Oct","Oct-Nov","Nov-Dec", "Dec-Jan"]
year = 1981
labels = []
while year < 2018:
    for month in months:
        labels.append(str(month) + " " + str(year))
    year += 1    
labels.append("Dec-Jan 2018")
labels.append("Jan-Feb 2018")
labels.append("Feb-Mar 2018")
labels.append("Mar-Apr 2018")
labels.append("Apr-May 2018")
labels.append("May-June 2018")
labels.append("June-July 2018")
labels.append("July-Aug 2018")
labels = ["Dec-Jan 1980"] + labels




labels
```

```python
Appl_json = {
    "numbers": changes, 
    "labels": labels
}
```

```python
import pymongo

# Create connection variable
conn = 'mongodb://localhost:27017'

# Pass connection to the pymongo instance.
client = pymongo.MongoClient(conn)

# Connect to a database. Will create one if not already available.
db = client.Apple_y_y

# Drops collection if available to remove duplicates
db.apple_y_o_y.drop()

# Creates a collection in the database and insert document
db.apple_y_o_y.insert_one(
    Appl_json
)
```
