```python
import pandas as pd
import requests
from datetime import datetime
```

```python
btc_info = requests.get("https://min-api.cryptocompare.com/data/histoday?fsym=BTC&tsym=USD&limit=207")
```

```python
btc_api_json = btc_info.json()
```

```python
import json
# json.pprint(btc_info.json())
# parsed = json.loads(btc_info.json())
print (json.dumps(btc_api_json, indent=4, sort_keys=True))
```

```python
high = []
low =[]
openP = []
close = []
timeS = []
volumefrom = []
volumeto = []
for x in btc_api_json["Data"]:
    high.append(x['high'])
    low.append(x['low'])
    openP.append(x["open"])
    close.append(x["close"])
    timeS.append(x["time"])
    volumefrom.append(x["volumefrom"])
    volumeto.append(x["volumeto"])
    
```

```python
btc_api_df = pd.DataFrame({
    "high": high, "low": low, "openP": openP, "close": close, "timeS": timeS, "volumefrom": volumefrom, "volumeto":volumeto
})
```

```python
btc_api_df = btc_api_df.rename(columns={'high': 'High ', 'openP': 'Open', 'close': 'Close', 'timeS':'Date',
                                        'low': "Low", "volumeto": "Volume"})
```

```python
# btc_api_df.head()
dTime = []
for row in btc_api_df["Date"]:
    x = datetime.utcfromtimestamp(row).strftime('%Y-%m-%d')
    dTime.append(x)
btc_api_df["Date"] = dTime
btc_api_df.head()
```

```python
btc_api_df = btc_api_df[["Date", "Open", "Close"]]
```

```python
btc_api_df.head()
```

```python
#### csv data
kaggle_set = pd.read_csv("bitcoin_price.csv")
```

```python
kaggle_set = kaggle_set[["Date", "Open", "Close"]]
kaggle_set.head()
```

```python
frames = [btc_api_df, kaggle_set]
bitcoin_df = pd.concat(frames)
```

```python
# bitcoin_df = bitcoin_df.reset_index()
bitcoin_df = bitcoin_df[["Date", "Open", "Close"]]
bitcoin_df.dtypes
```

```python
years = [2013, 2014, 2015, 2016, 2017, 2018]
firstHalf = []
secHalf = []
for year in years:
    Jan = bitcoin_df.loc[bitcoin_df['Date'] == f"{year}-01-01"]
    June = bitcoin_df.loc[bitcoin_df['Date'] == f"{year}-06-30"]
    July = bitcoin_df.loc[bitcoin_df['Date'] == f"{year}-07-01"]
    Dec = bitcoin_df.loc[bitcoin_df['Date'] == f"{year}-12-31"]
#     print(Dec["Date"])
    if Dec.empty:
        firstHalf.append(round((int(June["Close"])-int(Jan["Open"]))/int(Jan["Open"])*100))
    elif Jan.empty:
        secHalf.append(round((int(Dec["Close"])-int(July["Open"]))/int(July["Open"])*100))
    else:
        firstHalf.append(round((int(June["Close"])-int(Jan["Open"]))/int(Jan["Open"])*100))
        secHalf.append(round((int(Dec["Close"])-int(July["Open"]))/int(July["Open"])*100))
        
# print(firstHalf)
# print(secHalf)
print(Jan["Open"])
# print(June)
```

```python
bitCoinvBitcoin_Numbers = []
x = 0
while x < 5:
    bitCoinvBitcoin_Numbers.append(secHalf[x])
    bitCoinvBitcoin_Numbers.append(firstHalf[x])
    x += 1

bitCoinvBitcoin_Numbers    
```

```python
timeP = ["Second 2013", "First 2014", "Second 2014", "First 2015", "Second 2015",
        "First 2016", "Second 2016", "First 2017", "Second 2017", "First 2016"]
```

```python
bitCoin_json = {
    "numbers": bitCoinvBitcoin_Numbers, 
    "labels": timeP
}
```

```python
import pymongo

# Create connection variable
conn = 'mongodb://localhost:27017'

# Pass connection to the pymongo instance.
client = pymongo.MongoClient(conn)

# Connect to a database. Will create one if not already available.
db = client.bitcoin_y_o_y

# Drops collection if available to remove duplicates
db.bit_y_o_y.drop()

# Creates a collection in the database and insert document
db.bit_y_o_y.insert_one(
    bitCoin_json
)
```
